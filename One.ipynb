```python
import pandas as pd
import spacy
```

```python
helper = spacy.load('en_core_web_lg')
```

```python
df_example = pd.read_csv("../HW/Example_Technical_Skills.csv")
```

```python
df_example.head()
```

```python
eg1 = df_example['Technology Skills'][0]
eg1
```

```python
doc1 = helper(eg1)
```

```python
for token in doc1:
    print(token.pos_)
```

```python
num_of_ents = []
for i, data in df_example.itertuples():
    doc = helper(data)
    n = 0
    for token in doc.ents:
            n += 1
    num_of_ents.append(n)
```

```python
num_of_ents
```

```python
zero_nums = []
for i in range(0,len(num_of_ents)):
    if num_of_ents[i] == 0:
        zero_nums.append(i)
len(zero_nums)
```

```python
df_example['Technology Skills'][zero_nums]
```

```python
tech = 'technology'
tech_token = helper(tech)
```

```python
for data in df_example['Technology Skills'][zero_nums]:
    tokens = helper(data)
    for token in tokens:
#         print(f"{token} is similar to {tech_token} this much : {token.similarity(tech_token)}")
        print(f"{token.pos_}")
    print("\n\n")
```

```python
from spacy.matcher import PhraseMatcher
```

```python
matcher = PhraseMatcher(helper.vocab)
```

```python
phrase_list = df_example['Technology Skills'][zero_nums]
```

```python
phrase_patterns = [helper(text) for text in phrase_list]
```

```python
matcher.add('newproduct', None, *phrase_patterns)
```

```python
for i, data in df_example.itertuples():
    doc = helper(data)
    matches = matcher(doc)
    print(matches)
    print('\n')
```

```python
num_of_ents = []
for i, data in df_example.itertuples():
    doc = helper(data)
    n = 0
    for token in doc.ents:
            n+=1
    matches = matcher(doc)
    if matches:
        n+=1
    num_of_ents.append(n)
print(num_of_ents)
```

```python
len(num_of_ents)
```

```python
df_example['entities'] = num_of_ents
```
